```python
# Import ML Libraries

import pandas as pd
import numpy as np
from matplotlib import pyplot as pl 
import seaborn as sns
from pandas.core.internals.blocks import NumpyBlock
from spyder_kernels.utils.lazymodules import numpy
from sklearn.preprocessing import LabelEncoder
```

 # We are going to import all required matrics from sklearn
 from sklearn.metrics import confusion_matrix, accuracy_score, precision_score,recall_score, classification_report

```python
# Ground Truth Values

truth = ["Lion","Not a Lion","Lion","Lion",      "Lion", "Not a Lion", "Not a Lion", "Lion",       "Lion", "Not a Lion"] * 10

print(truth)
```

```python
# Predicted Values by a ML Model

prediction = ["Lion","Lion",      "Lion","Not a Lion","Lion", "Not a Lion", "Lion",       "Not a Lion", "Lion", "Lion"] * 10

print(prediction)
```

```python
# Use LabelEncoder to convert categories into numeric labels
label_encoder = LabelEncoder()
truth_binary = label_encoder.fit_transform(truth)
prediction_binary = label_encoder.transform(prediction)

print(truth_binary)
print(prediction_binary)
```

```python
# Calculate Confusion Matrix
from sklearn.metrics import confusion_matrix 
confusion_matrix = confusion_matrix(truth,prediction) 
print(confusion_matrix)
```

```python
#TP, FN, FP, TN = confusion_matrix(truth, prediction).ravel() - if the confusion_matrix is not declared above
TP, FN, FP, TN = confusion_matrix.ravel()
```

```python
print('True Positive(TP) = ', TP) 
print('False Positive(FP) = ', FP) 
print('True Negative(TN) = ', TN) 
print('False Negative(FN) = ', FN)
```

```python
# Displaying Heatmap with the Values of TP, FP, TN, FN
sns.heatmap(confusion_matrix, annot=True)
```

```python
# Displaying Heatmap with Labels TP, FP, TN, FN
labels = ['True Positive', 'False Negative',  'False Positive', 'True Negative']
labels = np.asarray(labels).reshape(2,2) 
sns.heatmap(confusion_matrix, annot=labels, fmt='', cmap='Reds')
```

```python
# Displaying Heatmap with Percentage Value of TP, FP, TN, FN
sns.heatmap(confusion_matrix/np.sum(confusion_matrix), annot=True, fmt='.2%',cmap='Greens')
```

```python
group_names = ['True Positive',   'False Negative', 'False Positive', 'True Negative']
group_counts = ["{0:0.0f}".format(value) for value in confusion_matrix.flatten()]
group_percentages = ["{0:.2%}".format(value) for value in confusion_matrix.flatten()/np.sum(confusion_matrix)]
labels = [f"{v1}\n{v2}\n{v3}" for v1, v2, v3 in zip(group_names,group_counts,group_percentages)]
labels = np.asarray(labels).reshape(2,2) 
sns.heatmap(confusion_matrix, annot=labels, fmt='', cmap='Purples')
```

```python
# Calculate Accuracy to show the total correct predictions
from sklearn.metrics import accuracy_score 
accuracy = accuracy_score(truth, prediction) 
print(accuracy)
```

```python
# Calculate Accuracy based on the Formula
#(all correct predictions/ total predictions) = TP + TN / TP + TN + FP + FN 
accuracy = (TP+TN) /(TP+FP+TN+FN)
print(accuracy)
```

```python
# Calculate Misclassification (opposite of Accuracy) 
#(all incorrect / total prditions) = FP + FN / TP + TN + FP + FN 
misclassification = (FP + FN) / (TP + TN + FP + FN) 
print(misclassification)
```

```python
# Calculate Precision between the two lists for each class (Lion and Not a Lion)
from sklearn.metrics import precision_score
precision = precision_score(truth, prediction, average=None, zero_division=0) 
print(precision)
```

```python
#Precision for a Lion class
precision1 = (TP) /(TP+FP) 
print(precision1)
```

```python
#Precision for "Not a Lion" class
precision2 = (TN) /(TN+FN) 
print(precision2)
```

```python
# Calculate Recall/Sensitivity between the two lists for each class (Lion and Not a Lion)
from sklearn.metrics import recall_score
recall = recall_score(truth, prediction, average=None, zero_division=0) 
print(recall)
```

```python
#Calculate Recall/Sensitivity for "Lion" class
recall1 = (TP) /(TP+FN) 
print(recall1)
```

```python
#Recall/Sensitivity for "Not a Lion" class
recall2 = (TN) /(TN+FP) 
print(recall2)
```

```python
# Calculate F1 between the two lists for each class (Lion and Not a Lion)
from sklearn.metrics import f1_score
F1 = f1_score(truth, prediction, average=None, zero_division=0) 
print(F1)
```

```python
# Calculate F1 between the two lists for each class (Lion and Not a Lion)
F1 = 2 * (precision * recall) / (precision + recall) 
print(F1)
```

```python
# Calculate F1 for "Lion" class
F1_1 = 2 * (precision1 * recall1) / (precision1 + recall1) 
print(F1_1)
```

```python
# Calculate F1 for "Not a Lion" class
F1_2 = 2 * (precision2 * recall2) / (precision2 + recall2) 
print(F1_2)
```

```python
# Generate the Complete Classification Report
from sklearn.metrics import classification_report 
classification_report = classification_report(truth, prediction) 
print(classification_report)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_curve, roc_auc_score, precision_recall_curve

# Use prediction_binary as features and truth_binary as labels
X = [[x] for x in prediction_binary]  # Convert prediction values into a two-dimensional feature array
y = truth_binary  # Labels

# Split the data into training and testing sets
trainX, testX, trainy, testy = train_test_split(X, y, test_size=0.4, random_state=2)

# Generate "No Skill" classifier predictions (predicting all class 0)
ns_probs = [0 for _ in range(len(testy))]

# Initialize multiple classifiers
classifiers = {
    'Logistic Regression': LogisticRegression(solver='lbfgs', max_iter=1000),
    'KNN': KNeighborsClassifier(n_neighbors=3),
    'Random Forest': RandomForestClassifier(n_estimators=100),
}

# Initialize ROC and Precision-Recall plots
pl.figure(figsize=(10, 6))
pl.title('Receiver Operating Characteristic (ROC) Curve')

# Train and evaluate each classifier
for name, clf in classifiers.items():
    clf.fit(trainX, trainy)  # Train the model
    pred_prob = clf.predict_proba(testX)[:, 1]  # Predict probabilities and get the probability for the positive class

    # Calculate the ROC curve
    fpr, tpr, _ = roc_curve(testy, pred_prob)
    roc_auc = roc_auc_score(testy, pred_prob)
    pl.plot(fpr, tpr, label=f'{name} (AUC = {roc_auc:.2f})')

# Plot the "No Skill" classifier's ROC curve
ns_fpr, ns_tpr, _ = roc_curve(testy, ns_probs)
pl.plot(ns_fpr, ns_tpr, linestyle='--', label='No Skill')

# Set the labels and title for the ROC plot
pl.xlabel('False Positive Rate')
pl.ylabel('True Positive Rate')
pl.legend()

# Display the ROC plot
pl.show()

# ----------- Add Precision-Recall Curve section --------------

# Initialize Precision-Recall plot
pl.figure(figsize=(10, 6))
pl.title('Precision-Recall Curve')

# Calculate Precision-Recall curve for each classifier
for name, clf in classifiers.items():
    pred_prob = clf.predict_proba(testX)[:, 1]  # Predict probabilities and get the probability for the positive class
    precision, recall, _ = precision_recall_curve(testy, pred_prob)
    pl.plot(recall, precision, label=f'{name}')

# Plot the "No Skill" classifier's Precision-Recall curve
ns_precision, ns_recall, _ = precision_recall_curve(testy, ns_probs)
pl.plot(ns_recall, ns_precision, linestyle='--', label='No Skill')

# Set the labels and title for the Precision-Recall plot
pl.xlabel('Recall')
pl.ylabel('Precision')
pl.legend()

# Display the Precision-Recall plot
pl.show()
```
