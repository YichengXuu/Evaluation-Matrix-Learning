```python
# Import all the required ML libraries

import pandas as pd
import numpy as np
from matplotlib import pyplot as pl 
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
```

# We are going to import all required matrics from sklearn
 from sklearn.metrics import confusion_matrix, accuracy_score, precision_score,recall_score, classificat

```python
#Ground True Values
truth = ["Cat", "Cat", "Cat", "Cat", "Cat", "Dog", "Dog","Dog","Dog","Dog", "Lion", "Lion", "Lion", "Lion", "Lion"] * 10

print(truth)
```

```python
# Predicted Values by a ML Model
prediction = ["Cat", "Lion", "Cat", "Dog", "Lion", "Dog", "Dog","Dog","Dog", "Lion", "Lion", "Lion", "Cat", "Lion", "Dog"] * 10

print(prediction)
```

```python
# Use LabelEncoder to convert categories into numeric labels
label_encoder = LabelEncoder()
y_true = label_encoder.fit_transform(truth) 
y_pred = label_encoder.transform(prediction) 

# Output the converted numeric arrays
print("Truth (Numeric):", y_true)
print("Prediction (Numeric):", y_pred)

# Output the class mapping corresponding to the labels
print("Class Mapping:", list(label_encoder.classes_))
```

```python
# Calculate Confusion Matrix
from sklearn.metrics import confusion_matrix 
confusion_matrix = confusion_matrix(truth,prediction) 
print(confusion_matrix)
```

```python
# Displaying Heatmap with the Values of Confusion Matrix
sns.heatmap(confusion_matrix, annot=True)
```

```python
# Displaying Heatmap with Labels TP, FP, TN, FN
labels = ['Cat - TP', 'Dog for Cat','Lion for Cat', 'Cat for Dog', 'Dog - TP','Lion for Dog', 'Cat for Lion', 'Dog for Lion', 'Lion - TP'] # use correct␣ 􏰀→apostrophes, chceck copy/paste mistake
labels = np.asarray(labels).reshape(3,3) 
sns.heatmap(confusion_matrix, annot=labels, fmt='', cmap='Reds')
```

```python
# Displaying Heatmap with Percentage Value of Confusion Matrix
sns.heatmap(confusion_matrix/np.sum(confusion_matrix), annot=True, fmt='.2%',cmap='Greens')
```

```python
group_names = ['Cat - TP', 'Dog for Cat','Lion for Cat', 'Cat for Dog', 'Dog - TP', 'Lion for Dog', 'Cat for Lion', 'Dog for Lion', 'Lion - TP']
group_counts = ["{0:0.0f}".format(value) for value in confusion_matrix.flatten()]
group_percentages = ["{0:.2%}".format(value) for value in confusion_matrix.flatten()/np.sum(confusion_matrix)]
labels = [f"{v1}\n{v2}\n{v3}" for v1, v2, v3 in zip(group_names,group_counts,group_percentages)]
labels = np.asarray(labels).reshape(3,3) 
sns.heatmap(confusion_matrix, annot=labels, fmt='', cmap='Purples')
```

```python
# Calculate Accuracy to show the total correct predictions
#(all correct predictions /total predictions)
from sklearn.metrics import accuracy_score 
accuracy = accuracy_score(truth, prediction) 
print(accuracy)
```

```python
# Calculate Misclassification (opposite of Accuracy)
#(all incorrect predictions /total predictions)
misclassification = 1 - accuracy
print(misclassification)
```

```python
# Calculate Precision for each class (Cat, Dog, Lion)
from sklearn.metrics import precision_score
precision = precision_score(truth, prediction, average=None, zero_division=0) 
print(precision)
```

```python
# Calculate Recall/Sensitivity for each class (Cat, Dog, Lion)
from sklearn.metrics import recall_score
recall = recall_score(truth, prediction, average=None, zero_division=0) 
print(recall)
```

```python
# Calculate F1 between the for each class (Cat, Dog, Lion)
from sklearn.metrics import f1_score
F1 = f1_score(truth, prediction, average=None, zero_division=0) 
print(F1)
```

```python
# Calculate F1 for each class (Cat, Dog, Lion)
F1 = 2 * (precision * recall) / (precision + recall) 
print(F1)
```

```python
# Generate the Complete Classification Report
from sklearn.metrics import classification_report 
classification_report = classification_report(truth, prediction) 
print(classification_report)
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_curve, precision_recall_curve, roc_auc_score
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.multiclass import OneVsRestClassifier

# Use prediction as the feature set and truth as the labels
X = np.array([[x] for x in y_pred])  # Use prediction values as features, ensure it is two-dimensional
y = y_true  # Use the original truth values as numeric labels

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=42)

# Initialize classifiers
classifiers = {
    'Logistic Regression': OneVsRestClassifier(LogisticRegression(solver='lbfgs', max_iter=1000)),
    'KNN': OneVsRestClassifier(KNeighborsClassifier(n_neighbors=3)),
    'Decision Tree': OneVsRestClassifier(DecisionTreeClassifier())
}

# Initialize the figure
pl.figure(figsize=(10, 6))

# Train and evaluate each classifier
for name, clf in classifiers.items():
    clf.fit(X_train, y_train)  # Train the model
    pred_prob = clf.predict_proba(X_test)  # Predict probabilities

    # ROC curve and AUC calculation
    fpr = {}
    tpr = {}
    n_class = len(set(y))  # Number of classes

    # Calculate the ROC curve for each class
    for i in range(n_class):
        fpr[i], tpr[i], _ = roc_curve(y_test, pred_prob[:, i], pos_label=i)
        pl.plot(fpr[i], tpr[i], label=f'{name} - Class {i} vs Rest')

# Plot ROC curve
pl.title('Multiclass ROC Curves for All Classifiers')
pl.xlabel('False Positive Rate')
pl.ylabel('True Positive Rate')
pl.legend(loc='best')
pl.show()

# Initialize the figure
pl.figure(figsize=(10, 6))

# Calculate Precision-Recall curves for each classifier
for name, clf in classifiers.items():
    clf.fit(X_train, y_train)  # Train the model
    pred_prob = clf.predict_proba(X_test)  # Predict probabilities

    # Precision-Recall curve
    precision = {}
    recall = {}

    # Calculate Precision-Recall curve for each class
    for i in range(n_class):
        precision[i], recall[i], _ = precision_recall_curve(y_test, pred_prob[:, i], pos_label=i)
        pl.plot(recall[i], precision[i], label=f'{name} - Class {i} vs Rest')

# Plot Precision-Recall curves
pl.title('Multiclass Precision-Recall Curves for All Classifiers')
pl.xlabel('Recall')
pl.ylabel('Precision')
pl.legend(loc='best')
pl.show()
```
